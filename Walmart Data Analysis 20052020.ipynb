```python
import numpy as np # import numpy
import pandas as pd # import pandad 
import matplotlib.pyplot as plt # import matplotlib
from matplotlib import style # import Style lib from matplot lib for Vizullzation 
plt.style.use('fivethirtyeight')
```

if you observe Variable 'Date' is having data type as String Now we have two options either Split this in same format or conver the data type in datetime and then extract the date and year. we will try both the option one by one and view the effect on the data.

```python
df_datetime = pd.to_datetime(df['Date']).dt.month # Converting string into date time data tyop 
#and then extract the monht from date 
```

```python
df_datetime.value_counts() # Check if month are propoer by viewing count of values available in the data
```

iIF we observe above data it show all Months are valid month

```python
df_year = pd.to_datetime(df['Date']).dt.year # For Sales data review we would have to extract the year also from the data
```

```python
df_year.value_counts() 
```

```python
df['Month']=df_datetime
df['Year']=df_year 
# now append this in to data file for further processing
```

```python
# Now Sort data with Stor, Month, year and Temp values by Pivot table
```

```python
df_store = pd.pivot_table(df,index=(df['Store'],df['Month'],df['Year'],df['Temperature']),aggfunc=np.sum).reset_index()
```

```python
df_store.head(2) # Check the Data for correction 
```

```python
df_store['Store'].value_counts()
```

```python
df_store['Year'].value_counts()
```

```python
df_store['Temperature'].value_counts()
```

```python
df_store.tail(2)
```

```python
df_store['Month'].value_counts().plot(kind='bar',figsize=(20,6),title=('Store Data')) 
```

```python
# plot Store wise Weekly Sales
plt.figure(figsize=(20,6)) 
plt.bar(df_store['Store'],df_store['Weekly_Sales'])
plt.title('Store Wise Weekly Sales')
plt.xlabel('Store')
plt.ylabel('Sales in $')
plt.tight_layout()
plt.show()
```

Above Plot show store wise Weekly Sales Fig

```python
# plot month wise Weekly Sales
plt.figure(figsize=(20,6)) 
plt.bar(df_store['Month'],df_store['Weekly_Sales'])
plt.title('Month Wise Weekly Sales')
plt.xlabel('Month')
plt.ylabel('Sales in $')
plt.tight_layout()
plt.show()
```

Fig Above Conclude that Max Sales had happned in month of December 

```python
# Temperature Effect on weekly Sales 
plt.figure(figsize=(20,6)) 
plt.scatter(df_store['Weekly_Sales'],df_store['Temperature'])
plt.title('Temperature Effect on Sales ')
plt.xlabel('Sales in $')
plt.ylabel('Temperature in Fahrenheit')
plt.tight_layout()
plt.show()
```

Observation:- 1. For Sales < 50000$ Templeratur doesnot have effect but as the Sales Value increases Temperatur is bounded between 20 to 60

```python
# Fule Price vs Temperature  
plt.figure(figsize=(20,6)) 
plt.scatter(df_store['Temperature'],df_store['Fuel_Price'])
plt.title('Temperature Effect on Fule Prices ')
plt.xlabel('Temperature in Fahrenheit')
plt.ylabel('Fule Prize in $')
plt.tight_layout()
plt.show()
```

```python
# Weekly Sales VS Unemployment 
plt.figure(figsize=(20,6)) 
plt.scatter(df_store['Weekly_Sales'],df_store['Unemployment'])
plt.title('Weekly Sales wise Unemployment ratio')
plt.xlabel('Weekly Sales')
plt.ylabel('Unemplyment')
plt.tight_layout()
plt.show()
```

This Show that Unemployment Rate is high if Sales is low Hence to over come unemplyment Sales needs to be increased 

```python
# Month VS Unemployment 
plt.figure(figsize=(20,6)) 
plt.bar(df_store['Month'],df_store['Unemployment'])
plt.title('Month wise Unemployment ratio')
plt.xlabel('Month')
plt.ylabel('Unemplyment')
plt.tight_layout()
plt.show()
```

This Shows that Max Unemployment is in the 6th Month and On an avg 13 people loss there job everymonth

```python
fig,ax = plt.subplots(1,1)
ax.hist(df_store['Temperature'], bins = [0,25,50,75,100])
ax.set_title("histogram of Temperature")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('Temperature ')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,6))
ax.barbs(df_store['Temperature'],df_store['Weekly_Sales'],label='Fule Price')
ax.barbs(df_store['Temperature'],df_store['Fuel_Price'],label='Unemployment')
```
